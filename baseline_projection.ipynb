```python
climatology = DS_date_range.groupby('time_counter.month').mean('time_counter')
anomalies = DS_date_range.groupby('time_counter.month') - climatology 
anomalies # compute climatology and anomalies
```

```python
DS_season = DS_date_range.groupby('time_counter.season').mean('time_counter') #seasonal mean
DS_season
```

```python
DS_winter = DS_season.sel(season='DJF') # select a specific season 
DS_winter
```

```python
DS_Season = DS_date_range.co2flux_cpl.groupby('time_counter.season').mean('time_counter')
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(9,5))
j = 0
for i, season in enumerate(('DJF', 'MAM', 'JJA', 'SON')):
    if season =='JJA':
        j += 1
        i = 0
    elif season =='SON':
        i = 1
        
    DS_Season.sel(season=season).plot.pcolormesh(
        ax=axes[i, j], vmin=-1, vmax=1, cmap='twilight', # class 'Qualitative'
        add_colorbar=True, extend='both')

for ax in axes.flat:
    ax.axes.get_xaxis().set_ticklabels([])
    ax.axes.get_yaxis().set_ticklabels([])
    ax.axes.axis('tight')
   
plt.tight_layout()
fig.suptitle('Seasonal CO2 fluxes', fontsize=16, y=1.05)
```
