```python
import netCDF4 as nc
from netCDF4 import Dataset
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import xarray as xr
import numpy as np
import pandas as pd
import calendar
import seaborn as sns
import seaborn_image as isns
```

```python
plt.style.use('seaborn-pastel')
isns.set_context(mode="notebook", fontfamily="Helvetica")
sns.set_context(font_scale=0.8)
```

```python
data_baseline = xr.open_dataset('/Volumes/UnionSine/data_transfer/ssp126/ALK_baseline_ensemble1_ssp126_1m.nc', decode_times=True).sel(deptht=slice(0, 100))
```

```python
data_oae = xr.open_dataset('/Volumes/UnionSine/data_transfer/ssp126/ALK_CDR_EU_coast_ssp126_1m.nc', decode_times=True).sel(deptht=slice(0, 100))
```

```python
sliced_data_baseline = data_baseline.where((-25 < data_baseline.nav_lon) & (data_baseline.nav_lon < 10) & (35 < data_baseline.nav_lat) & (data_baseline.nav_lat < 70), drop=True)
sliced_frame_baseline = sliced_data_baseline.sel(time_counter=slice('2090', '2100'))
```

```python
sliced_data_oae = data_oae.where((-25 < data_oae.nav_lon) & (data_oae.nav_lon < 10) & (35 < data_oae.nav_lat) & (data_oae.nav_lat < 70), drop=True)
sliced_frame_oae = sliced_data_oae.sel(time_counter=slice('2090', '2100'))
```

```python
sliced_data_baseline_plot = sliced_frame_baseline.ALK.mean(dim=('y', 'x', 'time_counter'))
sliced_data_oae_plot = sliced_frame_oae.ALK.mean(dim=('y', 'x', 'time_counter'))
```

```python
sliced_data_baseline_plot = sliced_data_baseline_plot.to_dataset()
sliced_data_oae_plot = sliced_data_oae_plot.to_dataset()
```

```python
plt.figure(figsize=(16, 3))
plt.plot(sliced_data_baseline_plot.deptht, sliced_data_baseline_plot.ALK, linewidth=1, label = 'Baseline')
plt.plot(sliced_data_oae_plot.deptht, sliced_data_oae_plot.ALK, linewidth=1, label = 'OAE')
plt.xlabel("Depth", fontsize=14)
plt.ylabel("Alkalinity in umol/kg", fontsize=14)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.title('EU Seasonal Average of Alkalinity from 2015 to 2100', fontname="Helvetica", fontsize=17, y=1.06)
plt.show()
```
