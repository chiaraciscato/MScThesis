```python
import netCDF4 as nc
from netCDF4 import Dataset
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import xarray as xr
import numpy as np
import pandas as pd
import calendar
import seaborn as sns
import seaborn_image as isns
```

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
plt.style.use('seaborn-pastel')
isns.set_context(mode="notebook", fontfamily="Helvetica")
sns.set_context(font_scale=0.8)
```

```python
alk_oae = xr.open_dataset('/Volumes/UnionSine/data_transfer/ssp126/ALK_CDR_EU_coast_ssp126_1m.nc', decode_times=True).sel(deptht=slice(0, 8))
```

```python
dic_oae = xr.open_dataset('/Volumes/UnionSine/data_transfer/ssp126/DIC_CDR_EU_coast_ssp126_1m.nc', decode_times=True).sel(deptht=slice(0, 8))
```

```python
sliced_alk = alk_oae.where((-25 < alk_oae.nav_lon) & (alk_oae.nav_lon < 10) & (35 < alk_oae.nav_lat) & (alk_oae.nav_lat < 70), drop=True)
```

```python
sliced_dic = dic_oae.where((-25 < dic_oae.nav_lon) & (dic_oae.nav_lon < 10) & (35 < dic_oae.nav_lat) & (dic_oae.nav_lat < 70), drop=True)
```

```python
start_date = "2090-01-01"
end_date = "2100-01-01"

alk_frame = sliced_alk["ALK"].sel(time_counter=slice(start_date, end_date))
```

```python
month_length_alk = alk_frame.time_counter.dt.days_in_month
alk_weighted = ((alk_frame * month_length_alk).resample(time_counter='QS-DEC').sum() / month_length_alk.resample(time_counter='QS-DEC').sum())
alk_weighted = alk_weighted.where(alk_weighted)
alk_weighted_season = alk_weighted.groupby('time_counter').mean(["y", "x"])
```

```python
start_date = "2090-01-01"
end_date = "2100-01-01"

dic_frame = sliced_dic["DIC"].sel(time_counter=slice(start_date, end_date))
```

```python
month_length_dic = dic_frame.time_counter.dt.days_in_month
dic_weighted = ((dic_frame * month_length_dic).resample(time_counter='QS-DEC').sum() / month_length_dic.resample(time_counter='QS-DEC').sum())
dic_weighted = dic_weighted.where(dic_weighted)
dic_weighted_season = dic_weighted.groupby('time_counter').mean(["y", "x"])
```

```python
colors = {3: "lightgreen", 6: "orange", 9: "brown", 12: "lightblue"}
seasons = {3: "Spring", 6: "Summer", 9: "Fall", 12: "Winter"}

f, ax = plt.subplots(figsize=(16, 3))
alk_weighted_season.plot(marker="o", markersize=7, markeredgewidth=0.5,
                        color="black", linewidth=0.5,
                        markerfacecolor="lightskyblue",
                        markeredgecolor="black", 
                        label = 'Alkalinity')

dic_weighted_season.plot(marker="o", markersize=7, markeredgewidth=0.5,
                        color="black", linewidth=0.5,
                        markerfacecolor="lightgreen",
                        markeredgecolor="black", 
                        label = 'DIC')
ax.set_title("DIC as a function of Alkalinity", fontname="Helvetica", fontsize=17, y=1.06)

plt.xlabel("Time", fontsize=14)
plt.ylabel("", fontsize=14)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

```python
alk_weighted = alk_weighted_season.to_dataframe('alk')
alk_weighted = alk_weighted.droplevel(1, axis=0)
alk_weighted['Month'] = pd.DatetimeIndex(alk_weighted.index).month
```

```python
conditions = [
    (alk_weighted['Month'] == 3),
    (alk_weighted['Month'] == 6),
    (alk_weighted['Month'] == 9),
    (alk_weighted['Month'] == 12)
]

values = ['Spring', 'Summer', 'Fall', 'Winter']

alk_weighted['Month'] = np.select(conditions, values)
```

```python
colors = dict(zip(alk_weighted.Month.unique(), ['paleturquoise', 'darkseagreen', 'gold', 'lightcoral']))
```

```python
fig, ax = plt.subplots(figsize=(16, 3))
ax.plot_date(alk_weighted.index, alk_weighted.alk, '-', color="black", linewidth=0.5, zorder=1)
ax.scatter(alk_weighted.index, alk_weighted.alk, color=alk_weighted.Month.map(lambda x: colors[x]), edgecolors='black',  linewidth=0.5, s=70)
legend = [mpatches.Patch(facecolor = value, edgecolor = "black", label = key, alpha = 1) for key, value in colors.items()]
ax.legend(handles = legend, bbox_to_anchor=(1, 1), loc='upper left')
ax.set_xlabel('Time', fontsize=14)
ax.set_ylabel('Alkalinity in umol/kg', fontsize=14)
ax.set_title("Weighted EU Seasonal Average of Alkalinity", fontname="Helvetica", fontsize=17, y=1.06)

pass
```

```python
dic_weighted = dic_weighted_season.to_dataframe('dic')
dic_weighted = dic_weighted.droplevel(1, axis=0)
dic_weighted['Month'] = pd.DatetimeIndex(dic_weighted.index).month
```

```python
conditions = [
    (dic_weighted['Month'] == 3),
    (dic_weighted['Month'] == 6),
    (dic_weighted['Month'] == 9),
    (dic_weighted['Month'] == 12)
]

values = ['Spring', 'Summer', 'Fall', 'Winter']

dic_weighted['Month'] = np.select(conditions, values)
```

```python
colors = dict(zip(dic_weighted.Month.unique(), ['paleturquoise', 'darkseagreen', 'gold', 'lightcoral']))
```

```python
fig, ax = plt.subplots(figsize=(16, 3))
ax.plot_date(dic_weighted.index, dic_weighted.dic, '-', color="black", linewidth=0.5, zorder=1)
ax.scatter(dic_weighted.index, dic_weighted.dic, color=dic_weighted.Month.map(lambda x: colors[x]), edgecolors='black',  linewidth=0.5, s=70)
legend = [mpatches.Patch(facecolor = value, edgecolor = "black", label = key, alpha = 1) for key, value in colors.items()]
ax.legend(handles = legend, bbox_to_anchor=(1, 1), loc='upper left')
ax.set_xlabel('Time', fontsize=14)
ax.set_ylabel('DIC in umol/kg', fontsize=14)
ax.set_title("Weighted EU Seasonal Average of DIC", fontname="Helvetica", fontsize=17, y=1.06)

pass
```

```python
colors = dict(zip(alk_weighted.Month.unique(), ['paleturquoise', 'darkseagreen', 'gold', 'lightcoral']))
colors = dict(zip(dic_weighted.Month.unique(), ['paleturquoise', 'darkseagreen', 'gold', 'lightcoral']))

fig, ax = plt.subplots(figsize=(16, 3))

ax.plot_date(alk_weighted.index, alk_weighted.alk, '-', color="black",  linewidth=0.5, zorder=1)
ax.scatter(alk_weighted.index, alk_weighted.alk, color=alk_weighted.Month.map(lambda x: colors[x]), marker='o', edgecolors='black',  linewidth=0.5, s=70)

ax.plot_date(dic_weighted.index, dic_weighted.dic, '-', color="black", linewidth=0.5, zorder=1)
ax.scatter(dic_weighted.index, dic_weighted.dic, color=dic_weighted.Month.map(lambda x: colors[x]), marker='o', edgecolors='black', linewidth=0.5, s=70)

legend = [mpatches.Patch(facecolor = value, edgecolor = "black", label = key, alpha = 1) for key, value in colors.items()]
ax.legend(handles = legend, bbox_to_anchor=(1, 1), loc='best')
ax.set_xlabel('Time', fontsize=14)
ax.set_ylabel('', fontsize=14)
ax.set_title("DIC as a function of Alkalinity", fontname="Helvetica", fontsize=17, y=1.06)

pass
```
