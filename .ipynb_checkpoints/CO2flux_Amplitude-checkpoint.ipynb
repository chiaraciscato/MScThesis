```python
plt.style.use('seaborn-pastel')
isns.set_context(mode="notebook", fontfamily="Helvetica")
sns.set_context(font_scale=0.8)
```

```python
ds_oae = xr.open_mfdataset('/Volumes/UnionSine/fco2/fco2_alk/*.nc')
```

```python
sliced_data = ds_oae.where((-25 < ds_oae.nav_lon) & (ds_oae.nav_lon < 10) & (35 < ds_oae.nav_lat) & (ds_oae.nav_lat < 70), drop=True)
```

```python
data_beginning = sliced_data.sel(time_counter=slice('2025', '2035'))
```

```python
fco2_max = data_beginning['fco2'].max('time_counter')
```

```python
fco2_min = data_beginning['fco2'].min('time_counter')
```

```python
fco2_amplitude = fco2_max - fco2_min
```

```python
fco2_amplitude = fco2_amplitude.to_dataset()
```

```python
plt.figure(figsize=(10, 4))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.add_feature(cfeature.LAND, zorder=100, facecolor='lightgrey', edgecolor='black')

fco2_amplitude.fco2.plot.pcolormesh(
    ax=ax, transform=ccrs.PlateCarree(), x="nav_lon", y="nav_lat", 
    add_colorbar=True, vmin = 0, vmax = 400
)
ax.coastlines()
ax.set_extent([-25, 10, 35, 70])
```

```python
data_ending = sliced_data.sel(time_counter=slice('2090', '2100'))
```

```python
fco2_max_end = data_ending['fco2'].max('time_counter')
```

```python
fco2_min_end = data_ending['fco2'].min('time_counter')
```

```python
fco2_amplitude_end = fco2_max_end - fco2_min_end
```

```python
fco2_amplitude_end = fco2_amplitude_end.to_dataset()
```

```python
plt.figure(figsize=(10, 4))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.add_feature(cfeature.LAND, zorder=100, facecolor='lightgrey', edgecolor='black')

fco2_amplitude_end.fco2.plot.pcolormesh(
    ax=ax, transform=ccrs.PlateCarree(), x="nav_lon", y="nav_lat", 
    add_colorbar=True, vmin = 0, vmax = 400
)
ax.coastlines()
ax.set_extent([-25, 10, 35, 70])
```
